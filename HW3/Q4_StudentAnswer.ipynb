```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
true_mu = 30 
sigma = 5   
N = 20


np.random.seed(0)
data = np.random.normal(true_mu, sigma, size=N)

#Guess here that student would answer faster in prior
mu0 = 20     
sigma0 = 10    

sigma_n_sq = 1 / (1/sigma0**2 + N/sigma**2)
mu_n = sigma_n_sq * (mu0/sigma0**2 + np.sum(data)/sigma**2)
```

```python
x = np.linspace(0, 60, 500)

prior = norm.pdf(x, mu0, sigma0)
posterior = norm.pdf(x, mu_n, np.sqrt(sigma_n_sq))

plt.plot(x, prior, label="Prior", linestyle="--")
plt.plot(x, posterior, label="Posterior")
plt.xlabel("Time (seconds)")
plt.ylabel("Density")
plt.title("Estimating Average Answer Time")
plt.legend()
plt.show()
```

```python
prior_samples = np.random.normal(mu0, sigma0, size=10000)
posterior_samples = np.random.normal(mu_n, np.sqrt(sigma_n_sq), size=10000)

plt.hist(prior_samples, bins=50, density=True, alpha=0.5, label="Prior")
plt.hist(posterior_samples, bins=50, density=True, alpha=0.5, label="Posterior")

plt.xlabel("Time (seconds)")
plt.ylabel("Density")
plt.title("Prior vs Posterior (Histogram)")
plt.legend()
plt.show()
```
